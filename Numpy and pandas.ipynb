### Method chaining and pipe

```python
df = pd.DataFrame({'A':['a','b','c','d'] * 25,
                   'B':np.arange(100),
                   'C':np.random.randn(100)})
df
```

```python
(df
 [lambda x: x.C > 0]
 .assign(D = df.B * df.C)
 .assign(E = lambda x: x.D - x.D.mean())
 .groupby('A')
 .E
 .sum())
```

```python
def a(x):
    return x[x['C'] > 0]
def b(x):
    return x.assign(D = x.B * x.C)
def c(x):
    return x.groupby('A')
def d(x):
    return x.sum()

(df.pipe(a)
   .pipe(b)
   .pipe(c)
   .pipe(d))
```
